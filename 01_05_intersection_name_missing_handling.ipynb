```python
import pandas as pd

# Load data from previous step
df = pd.read_parquet("data_01_04.parquet")

# Rename column and impute missing values
df.rename(columns={"Roundabout or junction": "Intersection name"}, inplace=True)
df["Intersection name"] = df["Intersection name"].fillna("not at intersection")

# Create distribution table
distribution_table = df["Intersection name"].value_counts(dropna=False).reset_index()
distribution_table.columns = ["Value", "Count"]
distribution_table["Percent"] = (distribution_table["Count"] / len(df) * 100).round(2)
```
