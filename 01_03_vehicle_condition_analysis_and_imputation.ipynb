```python
import pandas as pd

# Load data from previous step
df = pd.read_parquet("data_01_02.parquet")

# Analyze missingness
feature = "Vehicle condition"
missing_count = df[feature].isnull().sum()
missing_percent = missing_count / len(df) * 100
print(f"{feature}: {missing_count} missing ({missing_percent:.2f}%)")

# Raw value distribution
print("\nValue distribution before imputation:")
print(df[feature].value_counts(dropna=False))
```

```python
# Impute missing values as 'no visible defect'
df["Vehicle condition"] = df["Vehicle condition"].fillna("no visible defect")
```

```python
from IPython.display import display

# Distribution table
value_counts = df["Vehicle condition"].value_counts()
value_percent = df["Vehicle condition"].value_counts(normalize=True) * 100

distribution_table = pd.DataFrame({
    "Value": value_counts.index,
    "Count": value_counts.values,
    "Percent": value_percent.round(2).values
})

# Show in notebook
display(distribution_table)

# Save results
distribution_table.to_excel("vehicle_condition_table.xlsx", index=False)
df.to_parquet("data_01_03.parquet", index=False)
```
