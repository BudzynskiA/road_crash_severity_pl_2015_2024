```python
import pandas as pd

# Load data from previous step
df = pd.read_parquet("data_01_06.parquet")

# Check distribution of values in 'Under influence'
distribution_table = df["Under influence"].value_counts(dropna=False).reset_index()
distribution_table.columns = ["Value", "Count"]
distribution_table["Percent"] = (distribution_table["Count"] / len(df) * 100).round(2)

distribution_table
```

```python
# Translate existing values in 'Under influence'
translation_dict = {
    "Alkoholu": "Alcohol",
    "Innego środka": "Other substance",
    "Alkoholu; Innego środka": "Alcohol; Other substance",
    "Niebadany": "Not tested",
    "Innego środka; Niebadany": "Other substance; Not tested",
    None: "Unknown"
}

df["Under influence"] = df["Under influence"].map(translation_dict).fillna("Unknown")
```

```python
# Regenerate the translated distribution table
distribution_table = df["Under influence"].value_counts(dropna=False).reset_index()
distribution_table.columns = ["Value", "Count"]
distribution_table["Percent"] = (distribution_table["Count"] / len(df) * 100).round(2)

# Save to Excel
distribution_table.to_excel("under_influence_table.xlsx", index=False)

# Display the table
display(distribution_table)
```
