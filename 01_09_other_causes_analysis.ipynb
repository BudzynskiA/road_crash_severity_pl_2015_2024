```python
import pandas as pd

# Load full DataFrame
df = pd.read_parquet("data_01_08.parquet")

# Define the column to process
var = "Other causes"

# Step 1: merge duplicated labels (Polish)
df[var] = df[var].replace({
    "Z winy pasażera; Z winy pasażera": "Z winy pasażera"
})

# Step 2: translate Polish → English
translation_dict = {
    "None": "None",
    "Obiekty, zwierzęta na drodze": "Objects or animals on the road",
    "Nieustalone": "Undetermined",
    "Stan jezdni": "Road condition",
    "Inne": "Other",
    "Z winy pasażera": "Passenger's fault",
    "Niewłaściwy stan jezdni": "Poor road condition",
    "Niesprawność techniczna pojazdu": "Technical malfunction",
    "Utrata przytomności, śmierć kierującego": "Loss of consciousness / driver death",
    "Zabezpieczenie robót na drodze": "Road work protection",
    "Niezawiniona niesprawność techniczna pojazdu": "Unintentional technical malfunction",
    "Oślepienie przez inny pojazd lub słońce": "Glare from another vehicle or sun",
    "Pożar pojazdu": "Vehicle fire",
    "Nagłe zasłabnięcie kierującego": "Driver sudden illness",
    "Niesprawność techniczna pojazdu (od 11.2015)": "Technical malfunction (since 11.2015)",
    "Nieprawidłowo zabezp. roboty drogowe": "Improper road work protection",
    "Organizacja ruchu": "Traffic organization",
    "Utrata przytomności, śmierć kierującego (od 11...": "Loss of consciousness / driver death (since 11.2015)",
    "Nieprawidłowo działająca zapora kolej.": "Malfunctioning railway barrier",
    "Działanie sygnalizacji świetlnej": "Traffic light operation",
    "Nieprawidłowa organizacja ruchu": "Improper traffic organization",
    "Nieprawidłowo działająca zapora, rogatka": "Malfunctioning barrier",
    "Nieprawidłowo działająca sygn. świetlna": "Malfunctioning traffic light"
}
df[var] = df[var].replace(translation_dict)

# Step 3: impute 'None' with 'Not specified'
df[var] = df[var].replace("None", "Not specified")

# Save the updated full DataFrame
df.to_parquet("data_01_09.parquet", index=False)
print("File saved as: data_01_09.parquet")
```
