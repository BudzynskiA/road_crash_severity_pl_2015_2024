```python
import pandas as pd

# Columns of interest
cols = [
    "Vehicle ID",
    "Road markings",
    "Speed limit",
    "Weather conditions",
    "Road surface condition",
    "Crash site characteristics",
    "Road surface",
    "Area type",
    "Road type",
    "Traffic signal"
]

# Load selected columns
df = pd.read_parquet("data_01_22.parquet", columns=cols)

# Summary
for col in cols:
    print(f"=== {col} ===")
    print("Data type:", df[col].dtype)
    print("Number of unique non-null values:", df[col].nunique(dropna=True))
    print("Sample values:", df[col].dropna().unique()[:5])
    missing_count = df[col].isna().sum()
    missing_percent = df[col].isna().mean() * 100
    print(f"Missing: {missing_count} ({missing_percent:.2f}%)\n")
```
